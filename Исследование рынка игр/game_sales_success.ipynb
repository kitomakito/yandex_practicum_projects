### Находим и изучаем пропущенные значения в столбцах:

```python
#вычисляем долю пропусков по столбцам
df.isna().mean().sort_values(ascending=False)
```

<div style="border:solid green 2px; padding: 20px"> <b>Имеем критическое количество пропусков в столбцах critic_score (51%), rating (40%), user_score (40%).</b> Менее 1% - в столбце year_of_release, name, genre . <b>Пропуски в year_of_release, name и genre удалим, они не значительны. Пропуски в столбцах critic_score и  user_score оставим,</b> причину постараемся выяснить позднее; так как это оценачная характеристика, <b>будет некорректно, если мы ее заполним какой-то определённой величиной, это исказит картину, придётся всё заменить на отсутствие значений - nan. Пропуски в столбце rating заменим на "unknow"</b> так как этот столбец отвечает за возрастные ограничения по игре и само значение по-сути является категорией.</div>

```python
#по гистограмме видим, что основной диапазон - 70-78 очков; пики также в диапазонах 58-61 и 81-84
df.critic_score.hist(bins=30) 
plt.xlim(45,90)
plt.xlabel('Количество очков')
plt.title('Распределение оценок критиков')
plt.show()
print("Медиана столбца 'оценка критиков': ", df.critic_score.median())
```

 <div style="border:solid green 2px; padding: 20px"> По гистограмме "Оценки критиков" видим, что основной диапазон - 70-78 очков; пики также в диапазонах 58-61 и 81-84, медиана 71. В "Оценках пользователей" медиана - 7.5 (из 10), определена после очистки не числовых значений; большая часть оценок пользователей в диапазоне 8-9 баллов. Это высокие значения оценок, а так наша цель - подобрать популярные игры, завышать искусственно параметр, возможно влияющий на популярность и продажи нет смысла.</div>

```python
#Посмотрим на распределение значений в столбце "возрастная категория игр": 
df.rating.value_counts().plot(title = 'Рейтинг от ESRB (возрастная категория)', kind='bar', grid=True) 
plt.show()
```

Рейтинги ESRB дейтвительны в США и Канаде (в Европе и Японии свои рейтинговые системы: PEGI и CERO). 

RP ( «Rating Pending»») — «Рейтинг ожидается»: Продукт был отправлен в ESRB и ожидает присвоение рейтинга. Данный логотип используется только на рекламных презентациях и в демо-версиях игр до официального релиза. 

Рейтинг E (для всех) до 1988 года назывался K-A (для детей и взрослых), то есть это одна и та же категория.
<b>Заменим lдалее K-A на E</b>

```python
#Распределение доли пропусков в столбце rating по годам
df['nan_rating'] = df['rating'].isna()
df.groupby('year_of_release')['nan_rating'].agg('mean').plot(
    title='Доля пропусков в столбце "rating" по годам', kind='bar', grid=True, figsize=(10,3)
)
plt.show()
```

```python
#Распределение доли пропусков в столбце critic_score по годам
df['nan_critic_score'] = df['critic_score'].isna()
df.groupby('year_of_release')['nan_critic_score'].agg('mean').plot(
    title='Доля пропусков в столбце "critic_score" по годам', kind='bar', grid=True, figsize=(10,3)
)
plt.show()    
```

```python
#Распределение доли пропусков в столбце user_score по годам
df['nan_user_score'] = df['user_score'].isna()
df.groupby('year_of_release')['nan_user_score'].agg('mean').plot(
    title='Доля пропусков в столбце "user_score" по годам', kind='bar', grid=True, figsize=(10,3)
)
plt.show()
```

<div style="border:solid green 2px; padding: 20px"> 
o	Итак, мы наглядно видим, что после 1999 года произошёл резкий спад пропусков значений по всем трём параметрам (оценки критиков, пользователей, рейтинги игры); вероятно, причина в цифровизиции, переходу и росту информации, публикуемой непосредственно в Internet.</div>

```python
#удаляем пропуски, доля которых меньше 1%
df = df.dropna(subset=['year_of_release', 'name', 'genre']).reset_index(drop=True)
```

```python
#пропуски в столбце с возрастным рейтингом заполним значением "неизвестно"
df['rating'] = df['rating'].fillna('unknow')
```

```python
#замена 'K-A' на 'E' (устаревшее на современное значение):
df.loc[(df['rating'] == 'K-A'), 'rating'] = 'E'
```

```python
#проверяем
df['rating'].unique()
```

```python
df['user_score'].unique()
```

<b>Значение "tbd" - это оценка не известна, не определена. Смысла выделять её отдельно нет; оценка не известна и просто отсутствие оценки(nan) - одно и то же.</b>

```python
#заменяем значение "tbd" - оценка не известна - на "nan":
df.loc[df['user_score']=='tbd', 'user_score'] = float('nan')
```

## Проведите исследовательский анализ данных

### Выпуск игр по годам

```python
#Определим, сколько игр выпускалось в разные годы:
sales_by_year = df.groupby('year_of_release')['total_sales'].agg(['count', 'sum'])
sales_by_year.columns = ['Количество игр', 'Объём реализации']
```

```python
sales_by_year['Количество игр'].plot(kind='barh', figsize=(15,9)) #больше всего игр выпускалось с 2002 по 2011
plt.title('Количество выпущенных игр по годам')
plt.xlabel('Количество наименований игр, шт.')
plt.show()
```

```python
sales_by_year['Объём реализации'].plot(kind='bar', figsize= (15,5))
plt.title('Объём реализации, млн. копий')
plt.show()
```

<b>Видим, что не все данные таблицы важны для исследования, большая часть данных находится в диапазоне 2002-2011, пик в 2008-2009</b>

### Определяем платформы с наибольшими суммарными продажами

```python
#Строим столбчатый график изменений продаж по платформам:

df.groupby('platform')['total_sales'].agg(['sum', 'count']).sort_values(by='sum').plot(
    kind='bar', title='Суммарные продажи по платформам', figsize=(15,4)
)
plt.show()
```

По графику видим, что <b>больше всего продаж приходится на платформы PS, PS2, PS3, X360, Wii, DS.</b>

```python
#Смотрим распределение выбранных платформ по годам:

top_sales_by_platform = df.query('platform in ["PS2", "X360", "PS3", "Wii", "DS", "PS"]')
top_sales_by_plt_on_year = top_sales_by_platform.pivot_table(
    index='platform', columns = 'year_of_release', values = 'total_sales', aggfunc=['sum']
)
top_sales_by_plt_on_year.T
```

<b>Консоль DS была выпущена в 2004 году в Канаде, США и Японии, следовательно данные за 1985 год - аномалия. Удалим их</b>

### Находим характерный срок появления новых и исчезновения старых платформ

```python
#Строим график продаж игр на топ-платформах по годам:
top_sales_by_platform = df.query('platform in ["PS2", "X360", "PS3", "Wii", "DS", "PS"]')
plt.figure(figsize=(15,6))
sns.lineplot(data=top_sales_by_platform, x='year_of_release', y='total_sales', hue='platform', estimator = 'sum', ci=None)
plt.title('Жизненный цикл продаж самых игр на топ-платформах')
plt.show()
```

<div style="border:solid green 2px; padding: 20px"> Спад одной, появление новой игры на конкретной платформе занимает от 4 до 6 лет; сами платформы живут окола 10 лет.  Также на графике 'Количество выпущенных игр по годам' мы видим резкий разрыв между 2011 и 2012 годом, очевидна тенденция к снижению выпусков новых игр. Брать года до 2012 не имеет смысла из-за количественных изменений на рынке игр для платформ. 

Далее в анализе будем учитывать данные начиная с 2012 года до 2016.

Дополнительно отеметим, что самый долгий "жизненный цикл" у игр для PC (персональный компьютер); это связано с тем, что сама платформа не меняется, изменяются только операционные системы</div>

### Срезаем данные по актуальному периоду для прогнозирования

### Определяем популярные платформы, жанры

```python
#Определяем лидеров продаж(по платформам):
plt.figure(figsize=(15,6))
sns.lineplot(data=df_actual, x='year_of_release', y='total_sales', hue='platform', estimator = 'sum', style='platform', markers= True, ci=None, palette ='muted')
plt.title('Продажи по платформам с 2012 по 2015 года')
plt.show()
```

Самые "молодые" платформы - PS4, XOne, выглядят самыми перспективными, снижение продаж возможно связано с неполными данными, поскольку "запас" востребованности не исчерпан, платформам 2,5 года.
    
У остальных - низкие продажи, вялотекущий тренд на снижение.</b>

```python
#Постройте график «ящик с усами» по глобальным продажам игр в разбивке по платформам. Опишите результат.

plt.figure(figsize=(15, 8))
sns.boxplot(data=df_actual, x = 'platform', y = 'total_sales', palette='Set2')
plt.title('Диаграмма размаха для продаж по платформам')
plt.xlabel('Платформы')
plt.ylabel('Продажи по всем регионам, млн. копий')
plt.ylim(0, 2) #2.5
plt.show()
```

```python
#Общая картина
plt.figure(figsize=(20, 5))
sns.boxplot(data=df_actual, x = 'platform', y = 'total_sales', palette='Set2')
plt.title('Диаграмма размаха для продаж по платформам (общий вид)')
plt.xlabel('Платформы')
plt.ylabel('Продажи по всем регионам, млн. копий')
plt.show()
```

## Проверьте гипотезы

Проверим 2 гипотезы:

•	Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

•	Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные

```python
#Гипотеза 1
#H_0: Средние пользовательские рейтинги (user_score) пользователей "Xbox One" = средним user_score пользователей "PC"
#H_a: Средние пользовательские рейтинги пользователей "Xbox One" ≠ средним user_score пользователей "PC"

#подготавливаем массивы, содержащие выборки:
user_score_XOne = df_actual.loc[df_actual.platform=='XOne', 'user_score'].dropna()
user_score_PC = df_actual.loc[df_actual.platform=='PC', 'user_score'].dropna()
```

```python
#значение уровня значимости: alpha = 0.05

results = st.ttest_ind(
    user_score_XOne,
    user_score_PC,
    equal_var = False
) #нет причин предполагать что выборки одного размера (по графику 4.2 видна разница) 

alpha = 0.05 

# вывод значения p-value
print(results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha: #если значение меньше порогового
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
#Гипотеза 2: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) 
# и Sports (англ. «спортивные соревнования») разные.

#H_0: Средние пользовательские рейтинги (user_score) жанра "Action" = средним user_score жанра "Sports"
#H_a: Средние пользовательские рейтинги жанра "Action" ≠ средним user_score жанра "Sports"

#подготавливаем массивы, содержащие выборки:
user_score_Action= df_actual.loc[df_actual.genre=='Action', 'user_score'].dropna()
user_score_Sports= df_actual.loc[df_actual.genre=='Sports', 'user_score'].dropna()
```

```python
#задаём значение уровня значимости alpha = 0.05

results = st.ttest_ind(
    user_score_Action,
    user_score_Sports,
    equal_var = False
) ##нет причин предполагать что выборки одного размера игр жанра Action значительно больше 

alpha = 0.05 

# вывод значения p-value 
print(results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha: #если значение меньше порогового
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

<div class="alert alert-info"> <b>Вывод</b>: 
    
• Средние пользовательские рейтинги платформ Xbox One и PC одинаковые?

Нулевую гипотезу о равенстве среднего пользовательских оценок (user_score) пользователей "Xbox One" и пользователей "PC" <b>не получилось отвергнуть; средние пользовательские оценки с большей вероятностью равны.  </b>

• Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные?

<b>Нулевую гипотезу о равенстве среднего пользовательских оценок жанров Action и Sports мы отвергли.</b> Следовательно пользовательские рейтинги этих жанров да, разные.

<b>Нулевые гипотезы были сформулированы по принципу равенства средних двух генеральных совокупностей (для этого были взяты  выборка оценок пользователей Xbox One и выборка оценок пользователей PC; выборка оценок пользователей игр жанра Action и выборка оценок пользователей игр жанра Sports); следовательно альтернативная гипотеза выдвигалась как противоположная "нулевой" - о не равенстве средних двух генеральных совокупностей.</b>

Чтобы проверить гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам, нами был 
использован двусторонний t-тест Стьюдента (применён метод scipy.stats.ttest_ind()); 

Параметр equal_var указан в значении False, т.к. ранее в исследовательском анализе мы обнаружили перекосы в объёмах выборок (по популярным платформам и жанрам больше и оценок больше), у нет причин предполагать что выборки одного размера и их дисперсии будут равны.

<b>Для определения границ, по которой мы сможем понять, в каком моменте нам нужно отвергнуть нулевую гипотезу, а в каком нет, мы задали определённый уровнь значимости, который и является этим критическим значением для принятия решения по гипотезе: alpha = 0.05. </b>
</div>

<b>Самыми прибыльными жанрами(суммарные продажи больше 150 тыс копий) являются: action (экшн, действие), shooter (стрелялки), role-playing (ролевые), sport (спортивные)</b>; плохо продаваемыми жанрами являются: приключения, puzzle (головоломка), стратегия, сражения. Однако игры жанра action можно назвать рискованной ставкой - процент реализации составляет 48%

```python
df_actual.groupby('genre')['total_sales'].agg(['sum', 'count', 'mean']).sort_values(by='mean').head()
```
